```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot'
)
```

```python
import zipfile
import os

zip_file_path = '/content/traffic.csv.zip'
extract_dir = '/content/'
csv_file_name = 'traffic.csv' # Assuming the csv file inside the zip is named traffic.csv

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# Read the extracted CSV
df = pd.read_csv(os.path.join(extract_dir, csv_file_name))

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
total_sessions = df['linkid'].nunique()
print(total_sessions)
```

```python
pages = df.groupby('linkid')['event'].count()

bounce_rate = (pages==1).mean()*100

print(bounce_rate)
```

```python
session_duration_per_linkid = df.groupby('linkid')['date'].apply(lambda x: x.max() - x.min())
avg_duration = session_duration_per_linkid.mean()

print(avg_duration)
```

```python
landing = df.groupby('linkid').first()

landing['track'].value_counts().head(10).plot(
kind='bar',
figsize=(10,5))

plt.title("Top Landing Tracks")
plt.xlabel("Landing Track")
plt.ylabel("Sessions")
plt.show()
```

```python
exit_page = df.groupby('linkid').last()

exit_page['track'].value_counts().head(10).plot(
kind='bar',
figsize=(10,5))

plt.title("Top Exit Tracks")
plt.xlabel("Exit Track")
plt.ylabel("Sessions")
plt.show()
```

```python
sns.countplot(
data=df,
x='country', # Changed 'Traffic Source' to 'country'
order=df['country'].value_counts().index)

plt.xticks(rotation=45)
plt.title("Traffic Sources by Country") # Changed title to reflect the change
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))

top10 = df['country'].value_counts().head(10).index

sns.countplot(
    data=df,
    x='country',
    order=top10,
    palette='viridis'
)

plt.title("Top 10 Countries by Website Traffic", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Number of Visits")
plt.xticks(rotation=30, ha='right')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))

# Convert timedelta to minutes for plotting
plt.hist(session_duration_per_linkid.dt.total_seconds() / 60, bins=20)

plt.title("Session Duration Distribution") # Updated title for clarity
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```

Key Insights
1.Most visitors enter through a few landing pages.

2.Some exit pages have very high drop-offs.

3.Organic Search and Direct Traffic generate the highest number of sessions.

4.Mobile users contribute a significant portion of traffic.

5.verage session duration indicates overall engagement.

Recommendations for Alfido Tech
Improve high-bounce landing pages with clearer content and stronger calls to action.
Optimize top exit pages by adding internal links, related content, and conversion buttons.
Increase investment in high-performing referral sources such as Organic Search.
Improve the mobile user experience with faster page loading and responsive design.
Use personalized recommendations and targeted campaigns to encourage users to continue browsing and complete conversions.


Conclusion:

Website traffic analysis shows that user engagement depends heavily on landing page quality, traffic source, and session duration. By improving high-bounce pages, optimizing exit pages, enhancing the mobile experience, and focusing on high-performing traffic channels, Alfido Tech can increase user engagement and improve conversion rates.
